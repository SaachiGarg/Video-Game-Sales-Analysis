```python
import seaborn as sns

corr = df[['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']].corr()

sns.heatmap(corr, annot=True)
plt.show()
```

```python
top_games = df.sort_values('Global_Sales', ascending=False).head(20)

top_games[['Name','Platform','Genre','Global_Sales']]
```

```python
sorted_sales = df.sort_values('Global_Sales', ascending=False)
sorted_sales['cumulative'] = sorted_sales['Global_Sales'].cumsum()
sorted_sales['cumulative_pct'] = sorted_sales['cumulative'] / sorted_sales['Global_Sales'].sum()

sorted_sales[['Name','cumulative_pct']].head(50)
```

```python
genre_region = df.groupby('Genre')[['NA_Sales','EU_Sales','JP_Sales']].sum()

genre_region.plot(kind='bar', figsize=(12,6))
plt.show()
```

```python
avg_sales = df.groupby('Genre')['Global_Sales'].mean().sort_values(ascending=False)
avg_sales
```

```python
publisher_avg = df.groupby('Publisher')['Global_Sales'].mean().sort_values(ascending=False).head(10)
publisher_avg
```
